```python
import os
import pandas as pd

def load_titanic_train_data(titanic_train_data_path='datasets'):
    csv_path = os.path.join(titanic_train_data_path, 'train.csv')
    return pd.read_csv(csv_path)

titanic_train = load_titanic_train_data()
```

```python
titanic = titanic_train.drop("Survived", axis=1)
titanic_labels = titanic_train["Survived"].copy()
```

```python
from sklearn.base import BaseEstimator, TransformerMixin

class PassengerIdRemover(BaseEstimator, TransformerMixin):
    def __init__(self, remove_passenger_id=True):
        self.remove_passenger_id = remove_passenger_id
    def fit(self, X, y=None):
        return self
    def transform(self, X, y=None):
        return X.drop("PassengerId")
    
class NameRemover(BaseEstimator, TransformerMixin):
    def __init__(self, remove_name=True):
        self.remove_name = remove_name
    def fit(self, X, y=None):
        return self
    def transform(self, X, y=None):
        return X.drop("Name")
    
class TicketRemover(BaseEstimator, TransformerMixin):
    def __init__(self, remove_ticket=True):
        self.remove_ticket = remove_ticket
    def fit(self, X, y=None):
        return self
    def transform(self, X, y=None):
        return X.drop("Ticket")

class CabinRemover(BaseEstimator, TransformerMixin):
    def __init__(self, remove_cabin=True):
        self.remove_cabin = remove_cabin
    def fit(self, X, y=None):
        return self
    def transform(self, X, y=None):
        return X.drop("Cabin")
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder

rem_pipeline = Pipeline([
    ('passenger_id_remover', PassengerIdRemover()),
    ('name_remover', NameRemover()),
    ('ticket_remover', TicketRemover()),
    ('cabin_remover', CabinRemover()),
])

num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="median")),
    ('std_scaler', StandardScaler())
])

cat_pipeline = Pipeline([
    ('encoder', OneHotEncoder()),
    ('imputer', SimpleImputer(strategy="mode"))
])
```

```python
from sklearn.compose import ColumnTransformer

removable_attribs = ["PassengerId", "Name", "Ticket", "Cabin"]
num_attribs = ["Age", "Fare", "SibSp", "Parch"]
cat_attribs = ["Pclass", "Sex", "Embarked"]

full_pipeline = ColumnTransformer([
    ('rem', rem_pipeline, removable_attribs),
    ('num', num_pipeline, num_attribs),
    ('cat', cat_pipeline, cat_attribs)
])

titanic_prepared = full_pipeline.fit_transform(titanic)
```
